```python
#Recursion Task

def sum_nested_list(nested_list):  
     """
     Calculate the sum of all numbers in a nested list.
     This function takes a list that may contain integers and other nested lists.
     It recursively traverses the list and sums all the integers, no matter how deeply
     nested they are.
     Args:
     nested_list (list): A list that may contain integers or other lists of integers.
     Returns:
     int: The total sum of all integers in the nested list, including those in sublists
     .
     Example:
     >>> sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]])
     36
     >>> sum_nested_list([1, [2, 3], [4, [5]]])
     15
     """
     #docstring from the sample code
  
   

     totalsum=0
     for element in nested_list:
        if isinstance(element, list):
            totalsum+=sum_nested_list(element)
        else:
            totalsum+=element
         
     return totalsum
nested_list=[1, [2, [3, 4], 5], 6, [7, 8]]
total=sum_nested_list(nested_list)
print("the total sum of nested lists is", total)
     



```

```python
#Task2
def generate_permutations(s):
    """
    Generate all permutations of the given string.

    This function takes a string as input and returns a list of all possible 
    permutations of the string. The function uses recursion to generate the 
    permutations by fixing each character and recursively permuting the remaining characters.

    Args:
    s (str): The input string whose permutations are to be generated.

    Returns:
    list: A list containing all permutations of the input string.

    Example:
    >>> generate_permutations("abc")
    ['abc', 'acb', 'bac', 'bca', 'cab', 'cba']
    """

#docstring from the sample code
    if len(s)==1:
        return[s]

    permutations =[]

    for i in range(len(s)):
        
        remainingchars =s[:i] + s[i+1:]
        for perm in generate_permutations(remainingchars):
          
          permutations.append(s[i] + perm)
    return list(permutations)
s=generate_permutations("abc")

print("the total combination of the characters are:",s)
```

```python
 #Dynamic programming

#Task 1
def min_coins(coins, amount):
    """
    Calculates the minimum number of coins needed to make up a given amount.

    Uses dynamic programming to find the fewest coins required from a list of denominations.

    Args:
        coins (list): List of available coin denominations.
        amount (int): The target amount.

    Returns:
        int: The minimum number of coins needed, or -1 if it's not possible.
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    
    return dp[amount] if dp[amount] != float('inf') else -1
#example
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"the min coin needed to make amount {amount} is: {result}")

```

```python
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the Longest Common Subsequence (LCS) between two strings.

    The LCS is the longest sequence that appears in the same relative order in
    both strings, but not necessarily consecutively.

    Args:
        s1 (str): The first input string.
        s2 (str): The second input string.

    Returns:
        int: The length of the longest common subsequence between s1 and s2.
    """
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                 # If characters match, take the value from the diagonal and add 1
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                  # If they don't match, take the max value from the left or top cell
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

#example from the question
s1 = "abcde"
s2 = "ace"
result = longest_common_subsequence(s1, s2)
print(f"Length of the Longest Common Subsequence between '{s1}' and '{s2}': {result}")

```

```python
def knapsack(weights, values, capacity):#defining a function
    """
    Solves the 0/1 Knapsack problem using dynamic programming.

    This function calculates the maximum value that can be obtained by selecting
    items within the given weight capacity. Each item can either be included or
    excluded (0/1 Knapsack), and fractional selection is not allowed.

    Args:
        weights (list): A list of integers representing the weights of the items.
        values (list): A list of integers representing the values of the items.
        capacity (int): The maximum weight capacity of the knapsack.

    Returns:
        int: The maximum value that can be achieved with the given capacity.
    """
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    #filling the table row by row
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                 # Item can be included, take the maximum of including or excluding it
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                # Item can't be included, so take the value from the previous row
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]
#example from the question
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
result = knapsack(weights, values, capacity)
print(f"Maximum value that can be achieved with capacity {capacity}: {result}")
```
